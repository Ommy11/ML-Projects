```python
gender_data = df['gender'].value_counts()

explode = (0.1,0.1)
plt.figure(figsize = [14,8])
patches,texts,pcts = plt.pie(gender_data,
                             labels = gender_data.index,
                             colors = ['green','red'],
                             pctdistance = 0.65,
                             shadow = True,
                             startangle = 90,
#                              explode = explode,
                             autopct = '%1.1f%%',
                             textprops = {'fontsize':25,
                                         'color': 'black',
                                         'weight':'bold',
                                         'family':'serif'})

plt.setp(pcts,color='white')
hfont = {'fontname':'serif','weight':'bold'}
plt.title('gender data',size = 35,**hfont)

# centre_circle = plt.Circle((0,0),0.40,fc = 'white')
fig = plt.gcf()
# fig.gca().add_artist(centre_circle)
plt.show()
```

```python
age_group = [0, 20, 40, 60, 100]
labels = ['0-20','20-40','40-60','60-100']

#cresting a new column of age_group
df['age_group'] = pd.cut(df['age'],bins= age_group,labels = labels)
```

```python
#plot a histogram to determine the age_group
plt.figure(figsize = [8,6])
# df['age_group'].value_counts().sort_index().plot(kind='bar',color='red')
sns.countplot(df['age_group'])
plt.xlabel("Age group")
plt.ylabel("Count")
plt.title("Age Group wise distribution")
plt.show()
```
