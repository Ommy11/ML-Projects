```python
#visualising the given data

#creating the pairplot
sns.pairplot(df.drop("CustomerID",axis = 1))
```

```python
sns.countplot(x = "Gender",data = df)
plt.title("Customer Gender DIstribution")
plt.show()
```

```python
#segment customer by ages

sns.histplot(x='Age',data = df, color = "blue")
plt.title("Customer Age Distribution")
plt.show()
```

```python
#segment customer by incomes

sns.kdeplot(x = "Annual Income ($)",data = df,color = "purple",fill =True)
plt.title("IncomeDistribution")
```

```python
#determining the no. of customer by Profession

sns.countplot(data = df,x="Profession")
plt.title("Customer Profession Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
# determining by work experiences

sns.kdeplot(data =df,x="Work Experience",fill=True)
plt.title("Work Experience Distribution")
plt.show()
```

```python
# determine based  on family size

sns.countplot(x='Family Size',data = df)
plt.title("Family Size Distribution")
plt.show()
```

```python
# determining the Spending Score based on the gender

sns.violinplot(x="Gender",y="Spending Score (1-100)",data=df)
plt.title("Spending Score Distribution")
```
